def busqueda_lineal(lista, valor_buscado):

    for i in range(len(lista)):

        if lista[i] == valor_buscado:

            return i

    return -1

La b칰squeda binaria es un algoritmo m치s eficiente que la b칰squeda lineal, pero requiere que la lista est칠 ordenada. Divide la lista en dos partes y compara el valor buscado con el elemento central. Si no lo encuentra, descarta la mitad donde no podr칤a estar y repite el proceso con la otra mitad.

#Ejemplo de B칰squeda Binaria
Supongamos que tenemos una lista ordenada

[1,2,3,4,5,6,7,8,9] y queremos buscar el n칰mero 6.

Pseudoc칩digo:

Entrada: lista, valor_buscado

Proceso:
    establecer inicio = 0 y fin = longitud de la lista - 1

    Mientras inicio <= fin:

        calcular medio = (inicio + fin) / 2

        Si lista[medio] == valor_buscado:

            retornar medio

        Si lista[medio] < valor_buscado:

            inicio = medio + 1

        Sino:

            fin = medio - 1

    retornar -1 (si el valor_buscado no est치 en la lista)

Salida: 칤ndice del valor_buscado o -1

#C칩digo en Python:

def busqueda_binaria(lista, valor_buscado):

    inicio = 0
    fin = len(lista) - 1
    
    while inicio <= fin:
        medio = (inicio + fin) // 2
        if lista[medio] == valor_buscado:
            return medio
        elif lista[medio] < valor_buscado:
            inicio = medio + 1
        else:
            fin = medio - 1
    return -1

# Ejemplo de uso

lista = [1, 2, 3, 4, 5, 6, 7, 8, 9]

valor_buscado = 6

resultado = busqueda_binaria(lista, valor_buscado)
